```python
#문제 1

num = int(input('정수 입력 ==> '))

if num < 2:
    print(f'{num}은/는 소수가 아니다.')
else:
    for i in range(2, num):
        if num % i == 0:
            print(f'{num}은/는 소수가 아니다.')
            break
    else:
        print(f'{num}은/는 소수이다.')
```

```python
#문제 2

birth = input('생년월일 입력 ==> ')
month = birth[4] + birth[5]

rslt = ''

if len(birth) > 8:
    rslt = 'value error'
else:
    if month == '03' or month == '04' or month == '05':
        rslt = 'spring'
    elif month == '06' or month == '07' or month == '08':
        rslt = 'summer'
    elif month == '09' or month == '10' or month == '11':
        rslt = 'fall'
    elif month == '12' or month == '01' or month == '02':
        rslt = 'winter'
    else:
        rslt = 'value error'

print(rslt)
```

```python
#문제 3

plan = ['호텔 예약', '항공편 예약', '일정 작성', '명소 방문 예약']
time = [3, 2, 4, 1]

print('**초기 여행 계획 출력\n')

for i in range(len(plan)):
    print(f'여행 계획{i + 1}: {plan[i]} ==> {time[i]}시간')

print('----------------------------------------')
print('\n가장 긴 시간이 걸리는 계획과 가장 짧은 시간이 걸리는\n계획을 제외한 계획 목록')
max_index = time.index(max(time))
min_index = time.index(min(time))
remove_max = plan[max_index]
remove_min = plan[min_index]

plan.remove(remove_max)
plan.remove(remove_min)

print(plan)

print('----------------------------------------\n')
add_plan = input('추가할 여행 계획을 입력하세요: ')
add_time = int(input('계획 소요 시간을 입력하세요(시간 단위): '))
plan.append(add_plan)
time.append(add_time)

print('\n**수정된 여행 계획을 정렬하여 출력')
plan.sort()

print(plan)
```

```python
#문제 4

sentence = input('문장을 입력하시오: ')
original = sentence
secure_code = {'a': '1', 'e': '2', 'i': '3', 'o': '4', 'u': '5'}

sentence_list = list(sentence)
for i in range(len(sentence_list)):
    if sentence_list[i] in secure_code:
        sentence_list[i] = secure_code[sentence_list[i]]
secured_sentence = ''.join(sentence_list)

original_words = original.split()
secured_words = secured_sentence.split()
d = dict()

for i in range(len(secured_words)):
    d[original_words[i]] = secured_words[i]

print(d)
```

```python
#문제 5

def getBMI(tall, height):
    bmi = height / (tall * tall)
    return bmi

def getFat(bmi):
    rslt = ''

    if bmi < 18.5:
        rslt = '저체중'
    elif bmi < 25:
        rslt = '정상'
    elif bmi < 30:
        rslt = '과체중'
    else:
        rslt = '비만'
    return rslt

d = {'저체중': 0, '정상': 0, '과체중': 0, '비만': 0}
for i in range(5):
    tall = int(input('키 입력(cm단위): '))
    height = int(input('몸무게 입력(kg단위): '))
    bmi = getBMI(tall * 0.01, height)
    fat = getFat(bmi)
    d[fat] += 1

print('--- 학생 분포 ---')
print(d)
```
